```python
import requests
import json

ANKI_CONNECT_URL = "http://localhost:8765"

def invoke(action, **params):
    request = {"action": action, "version": 6, "params": params}
    response = requests.post(ANKI_CONNECT_URL, json=request).json()
    if len(response) != 2:
        raise Exception("Invalid response")
    if response.get("error"):
        raise Exception(response["error"])
    return response["result"]

# find all notes of the note type
note_ids = invoke("findNotes", query='deck:JP -is:suspended -is:buried -is:new')

# get fields for those notes
notes = invoke("notesInfo", notes=note_ids)

cloze_values = []

for note in notes:
    fields = note["fields"]
    if "Cloze" in fields:
        value = fields["Cloze"]["value"].strip()
        if value:
            cloze_values.append(value)

# optional: remove duplicates and sort
cloze_values = sorted(set(cloze_values))

for value in cloze_values:
    print(value)

#print length of the list
```

```python
#extract all the individual kanji from cloze_values and print them as a list
kanji_set = set()
for value in cloze_values:
    for char in value:
        if '\u4e00' <= char <= '\u9fff':  # Check if the character is a CJK Unified Ideograph
            kanji_set.add(char)
kanji_list = sorted(kanji_set)
print("\nUnique Kanji:")
for kanji in kanji_list:
    print(kanji)

#print length of the kanji list
print(f"Total unique kanji: {len(kanji_list)}")
```

```python
# unsuspend cards in the target deck that contain any of the extracted kanji (Kanji field only)
target_deck = "Japanese Heisig::Deck in progress"
target_field = "Kanji"

# find all cards in the deck
deck_card_ids = invoke("findCards", query=f"deck:\"{target_deck}\"")
card_infos = invoke("cardsInfo", cards=deck_card_ids)

# fetch note fields for the cards
note_ids = sorted({c.get("note") for c in card_infos if c.get("note") is not None})
note_infos = invoke("notesInfo", notes=note_ids)
notes_by_id = {n["noteId"]: n for n in note_infos}

kanji_set = set(kanji_list)
cards_to_unsuspend = []

for card in card_infos:
    note_id = card.get("note")
    note = notes_by_id.get(note_id)
    if not note:
        continue
    fields = note.get("fields", {})
    field = fields.get(target_field)
    if not field:
        continue
    value = (field.get("value") or "")
    if any(ch in value for ch in kanji_set):
        cards_to_unsuspend.append(card["cardId"])

# unsuspend matching cards
if cards_to_unsuspend:
    invoke("unsuspend", cards=cards_to_unsuspend)
    print(f"Unsuspended {len(cards_to_unsuspend)} cards")
else:
    print("No matching cards to unsuspend")
```
